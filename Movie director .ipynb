```python
# Import the libraries we need
import pandas as pd  # For working with data tables
import numpy as np   # For math operations
import matplotlib.pyplot as plt  # For creating charts

from wordcloud import WordCloud
```

```python
data = pd.read_csv('wiki_movie_plots.csv')
data.head(10)
```

```python
# select the specific director
director_name= input("Enter the director name : ")
director_data = data[data["Director"].str.contains(director_name,case=False)]

#Joining all plot description into single string
all_plots=" ".join(director_data["Plot"])

#createa world cloud
wordcloud= WordCloud(width=800,height= 400,background_color="white").generate(all_plots)
plt.figure(figsize=(10,6))
plt.imshow(wordcloud,interpolation="bilinear")
plt.title("Most common keywords in the movie plot")
plt.axis("Off")
plt.show()
```
